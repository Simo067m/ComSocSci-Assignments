```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
from IPython.display import display
```

```python
for f in glob("filtered_data/*"):
    if ".json" not in f: continue
    print(f)
    with pd.option_context('display.max_columns', None):
    # with pd.option_context('display.max_columns', None, 'display.max_colwidth', None):
        display(pd.read_json(f, orient='records', lines=True).head())
```

```python
# Load data
movies_df  = pd.read_json(r"filtered_data\final_movie_data.json", orient='records', lines=True)
cast_df    = pd.read_json(r"filtered_data\final_cast_data_with_movie_list.json", orient='records', lines=True)
crew_df    = pd.read_json(r"filtered_data\final_crew_data_with_movie_list.json", orient='records', lines=True)
company_df = pd.read_json(r"filtered_data\production_companies.json", orient='records', lines=True)
```

```python
# Return production company details by given company id
get_company = company_df.set_index("id").to_dict(orient="index")
# Production country as dict with Country Code as keys and Country name as values
get_country = pd.read_json(r"filtered_data\production_countries.json", orient='records', lines=True).set_index("Country Code").to_dict(orient="dict")['Country Name']
# Sponken language as dict with Language Code as keys and English name as values:
get_language = pd.read_json(r"filtered_data\spoken_languages.json", orient='records', lines=True).set_index("Language Code").to_dict(orient="dict")['English Name']
# Movie genres
get_genre = pd.read_json(r"filtered_data\movie_genres.json", orient="records", lines=True).set_index("genre_id").to_dict(orient="dict")['genre_name']
```

```python
# Check for duplicated id's (in each df, in pairs and all of them together)
has_dub = lambda DF: f"Has duplicates? {str(DF.size!=len(set(DF))).rjust(5)} | {len(set(DF)):>7}/{DF.size:<7} ({DF.size-len(set(DF)):>6} ~ {(DF.size-len(set(DF)))/DF.size*100:.2f}%)"

print(has_dub(movies_df.id))
print(has_dub(cast_df.id))
print(has_dub(crew_df.id))
print(has_dub(company_df.id))
print()
print(has_dub(pd.concat([movies_df.id,cast_df.id])))
print(has_dub(pd.concat([movies_df.id,crew_df.id])))
print(has_dub(pd.concat([cast_df.id,crew_df.id])))
```

```python
# Movie:
# budget, original_language, popularity, production_companies, production_countries, 
# release_date, revenue, runtime, spoken_languages, vote_average, 
# vote_count, combined_genres
# Cast:
# gender, known_for_department, popularity, movie_id, agg_characters
# Crew:
# gender, known_for_department, popularity, movie_id, agg_jobs, agg_departments
# Company:
# origin_country

def show_hist(col, title="Histogram", bins=50, logx=False, logy=False):
    cmean,cmedian = np.mean(col), np.median(col)
    fig, ax = plt.subplots(1,1,figsize=(20,8))
    # ax.hist(col, bins=bins, color='skyblue', edgecolor='black')
    if logx:
        bin_space = np.logspace(np.log10(max(col.min(),1)), # if 0, log would be not defined
                                np.log10(    col.max()   ),
                                bins)
    else:
        bin_space = np.linspace(col.min(),
                                col.max(),
                                bins)
    hist, edges = np.histogram(col,bins=bin_space)
    xlog = (edges[:-1]+edges[1:])/2      # midpoints between edges
    width = bin_space[1:]-bin_space[:-1]

    ax.bar(xlog,hist, width=width*0.85, color='skyblue', edgecolor='black')
    
    ax.set_title(title)
    ax.set_xlabel('Values')
    ax.set_ylabel('Count')
    ax.set_xscale('log' if logx else 'linear')
    ax.set_yscale('log' if logy else 'linear')
    if cmean > 10000: ax.axvline(cmean, color="blue", label=f"Mean: {np.mean(col):.3e}")
    else:             ax.axvline(cmean, color="blue", label=f"Mean: {np.mean(col):.3f}")
    if cmedian > 10000: ax.axvline(cmedian, color="green", label=f"Median: {np.median(col):.3e}")
    else:               ax.axvline(cmedian, color="green", label=f"Median: {np.median(col):.3f}")
    ax.legend()
    plt.show()
    
def show_list_count(col, apply_mapping = None, measure_str = "", top_count=15, bottom_count=15, use_barplot=True, logy=False):
    exploded_col = col.explode()
    if apply_mapping:
        exploded_col = exploded_col.map(apply_mapping)
    count_col = exploded_col.value_counts()
    if use_barplot:
        fig, ax = plt.subplots(1,1,figsize=(20,12))
        ax.barh(list(count_col.head(top_count).index), list(count_col.head(top_count).values), color='skyblue', edgecolor='black')
        # ax.set_xlabel('')
        ax.set_title(f"Top {top_count if top_count < count_col.size else ''}{measure_str}:")
        ax.set_xlabel('Count')
        ax.set_xscale('log' if logy else 'linear')
        plt.show()
    else:
        print(f"Top {top_count}{measure_str}:")
        display(count_col.head(top_count))
        print(f"\nBottom {bottom_count}{measure_str}:")
        display(count_col.tail(bottom_count))
```
